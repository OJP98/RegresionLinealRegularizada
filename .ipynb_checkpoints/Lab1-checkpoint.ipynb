## 1. Leer el archivo CSV

Importación de librerías

```python
import numpy as np
from matplotlib import pyplot as plt
from gradient_descent import gradient_descent
from linear_cost import linear_cost
from linear_cost_derivate import linear_cost_derivate
```

Leer el .csv

```python
admission = np.loadtxt('Admission_Predict.csv', delimiter="," , skiprows=1)
admission
```

### Análisis de variables

Las variables del dataset que pueden influir en el probablididad de aceptación (chance of admit) en la universidad son las siguientes:

* GRE Score
* TOEFL Score
* SOP
* LOR
* CGPA

Asignar variables desde el dataset

```python
# asignar variables de uso útil
gre = np.array(admission[:, 1])
toefl = np.array(admission[:, 2])
sop = np.array(admission[:, 3])
lor = np.array(admission[:, 4])
cgpa = np.array(admission[:, 6])

# El y se considera como el chance of admit
y = np.array(admission[:, 8])
TOTAL_ELEMENTS = len(y)

# Definir training set

    # Tamaño del dataset = 60% de los datos
TRAINNING_ELEMENTS = int(len(y)*0.6)
trainning = np.array(y[0:TRAINNING_ELEMENTS])
```

## 2. Ajustar un modelo lineal (polinómico)

### Comportamientos a notar

Las variables que parecen tener correlación con la probabilidad de admisión de la universidad, son las siguientes:
* GRE Score

Por lo tanto, se tomarán en cuenta estas variables para hacer la regresión.

### Creación de un modelo lineal

Crear el vector de unos y adjuntarlo a la matriz de datos, además de reajustar los datos resultado (y)

```python
# Se crea matriz de variables
x = np.array(admission[:, 6])

# Se crea vector de unos
ones = np.ones(TOTAL_ELEMENTS, dtype=int)

# se adjunta la matriz de datos con la columna de unos al inicio
X = np.column_stack((ones, x))

# Ajustar y para que solo sea una columna
trainning = trainning.reshape(TRAINNING_ELEMENTS,1)
y = y.reshape(TOTAL_ELEMENTS,1)
```

Obtener dimensiones m, n y proponer un theta inicial

```python
# Obtener m, n y proponer un theta (random)
m, n = X.shape
theta_0 = np.random.rand(n, 1)
theta_0 = np.array([[0.17912816],[0.92185169]])
theta_0
```

## 3. Regresión lineal regularizada

Utilizar la función de descenso al gradiente para entrenar al modelo y obtener variables

```python
# Obtener theta, los costos y las normales del gradientes
# con la función de descenso de gradiente
theta, costs, gradient_norms = gradient_descent(
    X[0:TRAINNING_ELEMENTS,:],
    trainning,
    theta_0,
    linear_cost,
    linear_cost_derivate,
    alpha=0.01,
    treshold=0.001,
    max_iter=10000,
    Lambda = 1,
)
```

Impresión de resultados y gráficos

```python
# Imprimir resultados
print('THETA:', theta)
plt.scatter(X[:, 1], y)
plt.plot(X[:, 1], np.matmul(X, theta), color='red')
plt.show()
```

Ya se poseen datos considerablemente adecuados para el modelo, por lo que el paso a seguir es ajustar el valor de lambda

## 4. Ajustando el valor  de lambda

Dado un alpha y un threshold quemados, se puede jugar con lambda para tener mejor rendimiento

```python
# Valores con los que funcionó el data set previo
_alpha = 0.01
_threshold = 0.001

theta, costs, gradient_norms = gradient_descent(
    X[0:TRAINNING_ELEMENTS,:],
    trainning,
    theta_0,
    linear_cost,
    linear_cost_derivate,
    alpha=_alpha,
    treshold=_threshold,
    max_iter=10000,
    Lambda = 0.001,
)
```

Impresión de resultados

```python
# Imprimir resultados
theta = np.array([[-0.89638356],[ 0.18818401]])
print('THETA:', theta)
plt.scatter(X[0:TRAINNING_ELEMENTS, 1], trainning)
plt.plot(X[0:TRAINNING_ELEMENTS, 1], np.matmul(X[0:TRAINNING_ELEMENTS], theta), color='red')
plt.show()
```

Una vez obtenido un entrenamiento adecuado, podemos quemar el valor de theta y lambda. Podemos empezar a hacer pruebas con cross validation

```python
_theta = theta
_lambda = 0.001
```

### Haciendo cross validation

```python
# Definir data frames
CROSS_VALIDATION_ELEMENTS = int(len(y)*0.8)

# Data frame del cross validation
cv = X[TRAINNING_ELEMENTS:CROSS_VALIDATION_ELEMENTS, :]
# Data frame del test set
test = X[CROSS_VALIDATION_ELEMENTS:TOTAL_ELEMENTS,:]
```

Una vez tenemos los data frames, podemos graficar nuestra regresión en el cross validation

```python
plt.scatter(cv[:, 1], y[TRAINNING_ELEMENTS:CROSS_VALIDATION_ELEMENTS, :])
plt.plot(cv[:, 1], np.matmul(cv, theta), color='red')
plt.show()
```

### Haciendo el test

```python
plt.scatter(test[:, 1], y[CROSS_VALIDATION_ELEMENTS:TOTAL_ELEMENTS,:])
plt.plot(test[:, 1], np.matmul(test, theta), color='red')
plt.show()
```

## 5. Análisis y hallazgos

```python
prediction = []
for i in range(TRAINNING_ELEMENTS):
    prediction.append(float(_theta[0] + _theta[1]*X[i,1]))

plt.hist(x = prediction, bins='auto', alpha=0.7, rwidth=0.85, label='predicción')
plt.hist(x = y[0:TRAINNING_ELEMENTS], alpha=0.7, rwidth=0.85, label='valor real')
plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('CGPA')
plt.ylabel('Frecuencia')
plt.title('Training Data "CGPA"')
plt.plot()
```

```python
prediction = []
for i in range(CROSS_VALIDATION_ELEMENTS,TOTAL_ELEMENTS):
    prediction.append(float(_theta[0] + _theta[1]*X[i,1]))

plt.hist(x = prediction, bins='auto', alpha=0.7, rwidth=0.85, label='predicción')
plt.hist(x = y[CROSS_VALIDATION_ELEMENTS:TOTAL_ELEMENTS], alpha=0.7, rwidth=0.85, label='valor real')
plt.grid(axis='y', alpha=0.75)
plt.legend(loc='upper right')
plt.xlabel('CGPA')
plt.ylabel('Frecuencia')
plt.title('Test Set "CGPA"')
plt.plot()
```

```python
percentages = []
for i in range(len(y)):
    answer = y[i]
    prediction = _theta[0] + _theta[1]*X[i,1]
    percentages.append(abs(float(((prediction-answer)/answer)*100)))

print("El porcentaje de error promedio sobre el test set del: {0:.2f}%".format(np.mean(percentages)))
```

## Uniéndo tres variables...

Para intentar crear un modelo más acertado, se juntarán las variables de GRE Score, TOEFL Score y CGPA

```python
X = np.hstack(
    (
        ones.reshape(-1,1),
        (gre.reshape(-1,1)**2)/100,
        (toefl.reshape(-1,1)**3)/1000,
        (cgpa.reshape(-1,1)**4)/10
    )
)

m, n = X.shape
theta_0 = np.random.rand(n, 1)

theta,_,_ = gradient_descent (
    X[0:TRAINNING_ELEMENTS,:],
    y[0:TRAINNING_ELEMENTS],
    theta_0,
    linear_cost,
    linear_cost_derivate,
    alpha=0.0000001,
    treshold=0.01,
    max_iter=10000000,
    Lambda = 3
)

# Este es el valor adecuado de theta. Si se desea probar con otro, comentar la siguiente línea:
theta = np.array([[ 4.64599842e-01],[-3.58451802e-04],[ 1.56895195e-04],[ 7.42187062e-04]])
```

```python
plt.scatter(X[0:TRAINNING_ELEMENTS,1], y[0:TRAINNING_ELEMENTS], color='red')
plt.scatter(X[0:TRAINNING_ELEMENTS,1], np.matmul(X[0:TRAINNING_ELEMENTS], theta), color='blue')
plt.title("Entrenamiento del conjunto de variables")
plt.show()
```

```python
_theta = theta

# Data frame del cross validation
cv = X[TRAINNING_ELEMENTS:CROSS_VALIDATION_ELEMENTS, :]
# Data frame del test set
test = X[CROSS_VALIDATION_ELEMENTS:TOTAL_ELEMENTS,:]

# Ahora se prueba con cross validation
plt.scatter(cv[:, 1], y[TRAINNING_ELEMENTS:CROSS_VALIDATION_ELEMENTS, :])
plt.scatter(cv[:, 1], np.matmul(cv, theta), color='red')
plt.title("Regresión lineal con el set de cross validation")
plt.show()
```

```python
# Ahora se prueba con el test set
plt.scatter(test[:, 1], y[CROSS_VALIDATION_ELEMENTS:TOTAL_ELEMENTS, :])
plt.scatter(test[:, 1], np.matmul(test, theta), color='red')
plt.title("Regresión lineal con el set de cross validation")
plt.show()
```

### Porcentaje de error del modelo

```python
percentages = []
for i in range(len(y)):
    answer = y[i]
    prediction = theta[0] + theta[1]*X[i,1] + theta[2]*X[i,2] + theta[3]*X[i,3]
    percentages.append(abs(float(((prediction-answer)/answer)*100)))

print("El porcentaje de error promedio sobre el test set del: {0:.2f}%".format(np.mean(percentages)))
```
